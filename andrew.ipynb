```python
artist_year_scores = df2.groupby(['name', 'year'])['year_end_score'].sum().reset_index()
artist_year_scores = artist_year_scores.sort_values(by = 'year')
artist_year_scores = artist_year_scores[artist_year_scores["name"] != "Various Artists"]

artist_year_scores
```

```python
# download artist_year_scores.csv
artist_year_scores.to_csv('artist_year_scores.csv', sep=',', index=False)
```
